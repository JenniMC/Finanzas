# Tratamiento

## Eliminacion de columnas vacias e innecesarias

Eliminar las ultimas 3 columnas que estaban vacias pero se habia quedado la suma en la tercer columna, cosa que no necesitaba

```python
data=data.iloc[:,:-3]
```

## Rellenar celdas nulas

```python
data.isnull().sum()
```

Llenar los nulos con cero, todos los nulos sin importar en que fila o columna esten

```python
data.fillna(0,inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

## Seleccionar rango de fechas a tratar

```python
data=data[data['Fecha']>='2023-01-01']
```

```python
data
```

```python
data=data[data['Fecha']<='2025-12-31']
```

```python
data
```

## Ordenar por fecha

```python
data=data.sort_values(by='Fecha')
```

```python
data
```

# Visualizacion

### Creacion de nuevas columnas



```python
data['Mes']=data['Fecha'].dt.month
data['Año']=data['Fecha'].dt.year
```

```python
data
```

```python
categorias_interes=['Ingreso']
data_ingresos=data[data['Categoria'].isin(categorias_interes)].copy()
```

```python
data_ingresos
```

```python
data.describe
```

### Aplicacion de abs

```python
data_ingresos['Cantidad']=data_ingresos['Cantidad'].abs()
```

### Agrupar las columnas

```python
resumen=data_ingresos.groupby(['Año','Mes'])['Cantidad'].sum().reset_index()
```

```python
resumen
```

## Graficar

```python
plt.figure(figsize=(12,6))
```

```python
años_a_graficar=[2023,2024,2025]
resumen=resumen[resumen['Año'].isin(años_a_graficar)]
sns.lineplot(data=resumen,x='Mes',y='Cantidad',hue='Año',marker='o',palette='viridis')
```

Aplicacion de formato

```python
sns.lineplot(data=resumen,x='Mes',y='Cantidad',hue='Año',marker='o',palette='viridis')
plt.title('Ingreso Mensual Total por Año')
plt.xticks(range(1,13),['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic'])
plt.ylabel('Ingreso Total')
plt.grid(True,alpha=0.3)
plt.show()
```
